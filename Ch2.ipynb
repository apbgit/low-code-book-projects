```python
import pandas as pd
url = 'https://data.cdc.gov/api/views/th8y-thx5/rows.csv?accessType=DOWNLOAD'
heart_df = pd.read_csv(url, index_col = 0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.violinplot(x='Data_Value', y='Stratification2', data=heart_df, hue='Stratification2', palette='Set2', legend=False)
plt.title('Heart Disease Mortality by Race/Ethnicity')
plt.xlabel('Data Value (per 100,000 population)')
plt.ylabel('Race/Ethnicity')
plt.show()
```
